```python
import json
```

```python
import tweepy
```

```python
consumer_key = '##'
consumer_secret = '##'
access_token = '##'
access_token_secret = '##'
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
tweets_list = tweepy.Cursor(api.search, q="#paralympics",
                            since='2021-08-24', 
                            until='2021-08-30',
                            tweet_mode='extended',
                            lang="en").items(700)
```

```python
output2 = []
for tweet in tweets_list:
    username=tweet.user.screen_name
    favourite_count = tweet.favorite_count
    retweet_count = tweet.retweet_count
    source = tweet.source
    line = {'id':tweet.id_str ,'username':username,'text' : tweet.full_text,'favourite_count':favourite_count, 'retweet_count' : retweet_count,
            'source': source,'location':tweet.user.location}
    output2.append(line)
```

```python
with open("paralympics-v5.json", "w") as outfile:
    json.dump(output2, outfile,indent=3)
```

```python
import pandas as pd
df=pd.DataFrame(output2)
```
